```python
path = '/content/task3.csv'
import pandas as pd
data = pd.read_csv(path)
data
```

```python
data= data.drop(columns = [data.columns[0]], axis = 1)
data
```

```python
l = data.columns.tolist()
l
```

```python
r = data[l[:3]]
r
```

```python
og_amount = data['Origination Amount']
rp = data.iloc[:, 1:].div(og_amount, axis =0)
rp
```

```python
rp_1 = rp['31.05.2019'].head(1) *2
import numpy as np
np.abs(rp_1 - rp['30.06.2019'].head(1))
```

```python
ei_rp = rp.copy()
ei_rp
```

```python
import numpy as np
for i in range(2, ei_rp.shape[1]):
  sum_prev = ei_rp.iloc[:, :i-1].sum(axis=1)
  ei_rp.iloc[:, i] = np.maximum(
      ei_rp.iloc[:, 1]*np.log(1+(1-i/30)*sum_prev),
      0
  )
ei_rp
```

```python
fc_cf = ei_rp.mul(og_amount, axis = 0)
fc_cf
```

```python
import matplotlib.pyplot as plt
import math
cols = 4
num_plots =len(fc_cf.columns)
rows = math.ceil(num_plots/cols)
fig, axes = plt.subplots(rows, cols, figsize=(20, rows*2.5), constrained_layout=True)
axes = axes.flatten()
for idx, i in enumerate(fc_cf):
  fc_cf[i].plot(ax = axes[idx], title = i)
  axes[idx].grid(True)
for i in range(num_plots, len(axes)):
  fig.delaxes[axes[i]]
plt.show()
```

```python
annual_dr = 0.025
mdr = (1+annual_dr)**(1/12) -1
df = [(1/(1+mdr))**i for i in range(1, fc_cf.shape[1]+1)]
df
```

```python
d = pd.DataFrame(df)
d.plot()
```

```python
pv = (fc_cf * df).sum().sum()
pv
```

```python
pv = round(pv, 2)
pv
```

```python
print(f'Portfolio Present Value: CHF {pv}')
```

```python
client_estimate = 84993122.67
client_estimate
```

```python
import numpy as np
abs_diff = np.abs(client_estimate - pv)
print(f'The absolute difference between the estimated and actual portfolio value: {round(abs_diff, 2)}')
```

```python
relative_diff = (abs_diff / client_estimate) * 100
print(f'Relative Difference: {round(relative_diff, 2)}%')
```
